## **Loading Data**

```python
import pandas as pd
```

```python
data = pd.read_csv('teams.csv')
data
```

### Removing unwanted coloums

```python
data = data[["team", "country", "year", "athletes", "age", "medals", "prev_medals"]]
data
```

## **Data Cleaning**

```python
data.isnull().sum()
```

```python
data[data['prev_medals'].isnull()]
```

```python
data = data.dropna()
data
```

## **Splitting Data**

```python
train = data[data['year'] < 2012].copy()
test = data[data['year'] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

## **Buliding Our Model**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ['athletes', 'prev_medals']
target = 'medals'
```

```python
reg.fit(train[predictors], train['medals'])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions
```

### **Scailing the Model**

```python
test ['predictions'] = predictions
test
```

```python
test.loc[test['predictions'] < 0, 'predictions'] = 0
test ['predictions'] = test['predictions'].round()
test
```

### **Measuring Error**

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test['medals'], test['predictions'])
error
```

```python
test.describe()['medals']
```

### **Analyzing Our Model**

```python
test[test['team'] == 'IND']
```

```python
test[test['team'] == 'USA']
```

```python
errors = (test['medals'] - test['predictions']).abs()
errors
```

```python
error_by_team = errors.groupby(test['team']).mean()
error_by_team
```

```python
medals = test['medals'].groupby(test['team']).mean()
medals
```

```python
error_rate = error_by_team / medals
error_rate
```
